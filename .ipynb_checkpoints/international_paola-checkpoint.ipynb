```python
#setting up 
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress 
from pprint import pprint 

#importing csv data variables 
international_drug_seizures_path = "seizures_1.csv"
world_data_path = "worlddata.csv"

drug_seizures = pd.read_csv(international_drug_seizures_path)
world_data = pd.read_csv(world_data_path)

#marge data
wdxds = pd.merge(drug_seizures, world_data, how='outer', on='country')
wdxds
```

```python
#creating a data frame for drug seizures by country and type of drug 
cannabis = wdxds['Drug Group'].groupby(wdxds['country'])
opioids = wdxds['Drug Group'].groupby(wdxds['country'])

summary = pd.DataFrame({"Cannabis": cannabis, "Opioids" : opioids })
summary
```

```python
#bar plot of drug seizures by country 
drug_count = wdxds['Drug Group'].value_counts()
drug_bar = drug_count.plot.bar(color='steelblue')
plt.ylabel("Number of Drug Seizures (1970-2020)")
```
